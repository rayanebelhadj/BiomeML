```python
#on définit les valeurs positives et négatives
diab_true = ['Diagnosed by a medical professional (doctor, physician assistant)']
diab_false = [ 'I do not have this condition']
```

## Étape 3: faire un subset de la dataframe de diabète
Le but ici est de se faire un sous-ensemble de diabétiques et de non-diabétiques qui puissent être comparables pour pouvoir servir de benchmark. On va donc échantillonner des échantillons non-diabétiques avec une distribution d'âges similaires à celle des échantillons diabétiques

```python
no_diabetes = []
t2d = []
t2d_ages = []
for ind in range(len(lst_db)):
    if lst_db[ind] in diab_true:
        if list_type[ind]=='Type II diabetes':
            t2d.append(list_samples[ind])
            t2d_ages.append(lst_ages[ind])
    elif lst_db[ind] in diab_false:
        no_diabetes.append(list_samples[ind])
```

```python
len(no_diabetes)
```

```python
len(t2d)
```

```python
t2d_ages
```

```python
df_t2d = df[(df["diabetes"]=='Diagnosed by a medical professional (doctor, physician assistant)') & (df["diabetes_type"]=="Type II diabetes")]
```

```python
df_not2d = df[(df["diabetes"]=='I do not have this condition')]
```

```python
df_t2d
```

```python
target_counts = df_t2d.groupby(['age_cat', 'bmi_cat']).size()

# Step 2: Filter full df to rows with same A value
subset_df = df_not2d

# Step 3: Sample rows according to target_counts
samples = []

for (b_val, c_val), count in target_counts.items():
    matching_rows = subset_df[(subset_df['age_cat'] == b_val) & (subset_df['bmi_cat'] == c_val)]
    
    if len(matching_rows) >= count:
        sampled = matching_rows.sample(n=count, replace=False)
    else:
        # Not enough rows — you can choose to skip, or sample with replacement
        sampled = matching_rows.sample(n=count, replace=True)  # or just `continue` to skip
    samples.append(sampled)

# Step 4: Combine and shuffle
final_sample = pd.concat(samples).sample(frac=1).reset_index(drop=True)
```

```python
final_sample.to_csv("AGP_T2Dcontrol.metadata.txt", sep="\t", index=False)
```

## Étape 4: filtrage des données
Avec les métadonnées, on a une super grosse table BIOM, qui contient toutes les séquences et tous les échantillons (chaque colonne est un échantillon et chaque rangée est une séquence). La table s'appelle AGP.data.biom.qza, et on va la filtrer avec nos sous-ensembles de métadonnées.

```python
#first, filter biom table to only the features in df_t2d
```

```python
!qiime feature-table filter-samples --i-table AGP.data.biom.qza --m-metadata-file ../AGP_T2D.metadata.txt --o-filtered-table AGP.data.biom.filtered_samples.T2D.qza
```

```python
#then, remove all features with low count
```

```python
!qiime feature-table filter-features-conditionally --i-table AGP.data.biom.filtered_samples.T2D.qza --o-filtered-table AGP.data.biom.filtered_samples.T2D.filtered_features.qza --p-abundance 0.01 --p-prevalence 0.01
```

```python
#then, export the table and get the sequences and map to greengenes
```

```python
!qiime tools export --input-path AGP.data.biom.filtered_samples.T2D.filtered_features.qza --output-path FILTERED_T2D_BIOM 
!cd FILTERED_T2D_BIOM
!biom convert -i feature-table.biom -o AGP.data.fiom.filtered.T2D.tsv --to-tsv
!biom table-ids -i feature-table.biom --observations |     awk '{ print ">" $1 "\n" $1 }' > AGP_seqs.fa
```
